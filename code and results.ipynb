```python
### Part 1: Imports and Core Configuration


!pip install monai optuna -q
import os
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"
import os, warnings, logging
warnings.filterwarnings("ignore")
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3" 
for lib in ["monai", "absl", "torch", "tensorflow", "matplotlib"]:
    logging.getLogger(lib).setLevel(logging.ERROR)
import gc
import sys
import glob
import time
import random
import json
import copy
import shutil
import tempfile
import numpy as np
import nibabel as nib
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from scipy.ndimage import zoom
from sklearn.model_selection import train_test_split
import cv2
import optuna


if torch.cuda.is_available():
    torch.backends.cudnn.benchmark = False
    torch.backends.cudnn.deterministic = True
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.backends.cudnn.allow_tf32 = True
    # PyTorch 2.0+
    try:
        torch.set_float32_matmul_precision("high")
    except Exception:
        pass


# MONAI is used for specific loss and metric calculations.
try:
    from monai.losses import HausdorffDTLoss
    from monai.metrics import HausdorffDistanceMetric
except ImportError:
    print("MONAI not found. Please ensure it is installed to use this script.")
    sys.exit()

class Config:
    BASE_DIR = '/kaggle/input/brats20-dataset-training-validation/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData'
    PROCESSED_DIR = '/kaggle/working/processed_brats2020_2d_9ch'
    OUTPUT_DIR = '/kaggle/working/output_final'

    DEVICE = "cpu"
    AMP_ENABLED = False
    NUM_WORKERS = 0
    PERSISTENT_WORKERS = False
    PREFETCH_FACTOR = None

    RANDOM_STATE = 42

    IMG_SIZE = 128
    BATCH_SIZE = 24
    LEARNING_RATE = 1e-4
    WEIGHT_DECAY = 1e-5
    GRAD_CLIP_NORM = 1.0
    DROPOUT_PROB = 0.2

    HARD_NEGATIVE_COUNT = 4

    EPOCHS_STAGE1 = 25
    EPOCHS_STAGE2 = 35

    TUNE_EPOCHS_STAGE1 = 10
    TUNE_EPOCHS_STAGE2 = 10
    N_TUNING_TRIALS = 5

    PSEUDO_BINARIZE_THR = 0.6
    PSEUDO_MIN_SIZE = 25
    CONFIDENT_NEGATIVE_THR = 0.05

    STUDENT_TEMP = 0.75
    STUDENT_BOOTSTRAP_BETA = 0.8

    S1_BCE_WEIGHT, S1_DICE_WEIGHT, S1_HAUSDORFF_WEIGHT = 0.4, 0.4, 0.2
    S3_BCE_WEIGHT, S3_DICE_WEIGHT, S3_HAUSDORFF_WEIGHT = 0.5, 0.5, 0.0

    SCHEDULER_PATIENCE = 3
    SCHEDULER_FACTOR = 0.5

# Setup for Performance and Reproducibility 
def setup_environment(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)


setup_environment(Config.RANDOM_STATE)
os.makedirs(Config.OUTPUT_DIR, exist_ok=True)
print(f"Configuration loaded. Using device: '{Config.DEVICE}'")
print(f"All generated artifacts will be saved to: '{Config.OUTPUT_DIR}'")




# Part 1.5: Visualize Raw MRI Modalities



print("\n--- Visualizing Raw MRI Modalities ---")

def get_vis_file_path(base_dir, patient_id, suffix):
    """A helper to find file paths, handling a known dataset anomaly and .nii vs .nii.gz files."""
    if suffix == 'seg' and patient_id == "BraTS20_Training_355":
        alt_path = os.path.join(base_dir, patient_id, "W39_1998.09.19_Segm.nii")
        if os.path.exists(alt_path): return alt_path
    path = os.path.join(base_dir, patient_id, f'{patient_id}_{suffix}.nii')
    return path if os.path.exists(path) else path + '.gz'

# Find a sample patient to display
all_patient_dirs_vis = sorted([d for d in glob.glob(os.path.join(Config.BASE_DIR, 'BraTS20_Training_*')) if os.path.isdir(d)])
if all_patient_dirs_vis:
    sample_patient_id = os.path.basename(all_patient_dirs_vis[0])  # keep your selection
    try:
        flair_vol_vis = nib.load(get_vis_file_path(Config.BASE_DIR, sample_patient_id, 'flair')).get_fdata()
        t1ce_vol_vis = nib.load(get_vis_file_path(Config.BASE_DIR, sample_patient_id, 't1ce')).get_fdata()
        t2_vol_vis   = nib.load(get_vis_file_path(Config.BASE_DIR, sample_patient_id, 't2')).get_fdata()
        seg_vol_vis  = nib.load(get_vis_file_path(Config.BASE_DIR, sample_patient_id, 'seg')).get_fdata()

        slice_to_show = flair_vol_vis.shape[2] // 2
        fig, axes = plt.subplots(1, 4, figsize=(18, 6))
        plt.suptitle(f"MRI Modalities for Patient: {sample_patient_id} (Slice {slice_to_show})", fontsize=20)

        axes[0].imshow(flair_vol_vis[:, :, slice_to_show], cmap='bone'); axes[0].set_title("Flair", fontsize=20); axes[0].axis('off')
        axes[1].imshow(t1ce_vol_vis[:, :, slice_to_show], cmap='bone'); axes[1].set_title("T1ce", fontsize=20); axes[1].axis('off')
        axes[2].imshow(t2_vol_vis[:, :, slice_to_show], cmap='bone'); axes[2].set_title("T2", fontsize=20); axes[2].axis('off')
        axes[3].imshow(seg_vol_vis[:, :, slice_to_show], cmap='bone')
        axes[3].set_title("Seg Mask (binary)", fontsize=20); axes[3].axis('off')
        
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        # NEW: save high-quality image
        raw_out = os.path.join(Config.OUTPUT_DIR, f"raw_modalities_{sample_patient_id}_s{slice_to_show}.png")
        plt.savefig(raw_out, dpi=500, bbox_inches='tight')
        print(f"Saved: {raw_out}")
        plt.show()
    except Exception as e:
        print(f"Could not generate visualization for patient {sample_patient_id}: {e}")
else:
    print("Could not find patient directories to generate visualization.")



### Part 2: Core Components (Models, Losses, Datasets, etc.)

def get_file_path(base_dir, patient_id, suffix):
    """Constructs the file path for a given patient and modality."""
    if suffix == 'seg' and patient_id == "BraTS20_Training_355":
        alt_path = os.path.join(base_dir, patient_id, "W39_1998.09.19_Segm.nii")
        if os.path.exists(alt_path): return alt_path
    path = os.path.join(base_dir, patient_id, f'{patient_id}_{suffix}.nii')
    return path if os.path.exists(path) else path + '.gz'


def dl_kwargs(cfg):
    use_cuda = torch.cuda.is_available()
    kw = dict(
        num_workers=cfg.NUM_WORKERS,
        pin_memory=use_cuda,
    )
    if cfg.PERSISTENT_WORKERS and cfg.NUM_WORKERS > 0:
        kw["persistent_workers"] = True
    if cfg.NUM_WORKERS > 0 and cfg.PREFETCH_FACTOR is not None:
        kw["prefetch_factor"] = cfg.PREFETCH_FACTOR
    return kw


def crop_to_brain(image_slice, optional_padding=10):
    """Crops a 2D or 3D image to the non-zero brain region."""
    is_2d = (image_slice.ndim == 2)
    img3 = image_slice[..., None] if is_2d else image_slice
    # Find bounding box of non-zero pixels
    nz_indices = np.argwhere(np.any(img3 > 1e-6, axis=-1))
    if nz_indices.size == 0: return image_slice, None # No brain found
    y0, x0 = nz_indices.min(axis=0)
    y1, x1 = nz_indices.max(axis=0)
    # Add padding
    y0, x0 = max(0, y0 - optional_padding), max(0, x0 - optional_padding)
    y1, x1 = min(img3.shape[0], y1 + optional_padding), min(img3.shape[1], x1 + optional_padding)
    out = img3[y0:y1, x0:x1]
    return (out[..., 0] if is_2d else out), (y0, x0, y1, x1)

def stack_neighbors(flair, t1ce, t2, z_index):
    """Stacks a slice z with its neighbors from three MRI modalities to create a 9-channel input."""
    z_max = flair.shape[2]
    channels = []
    for dz in (-1, 0, 1): # Previous, current, and next slice
        z = int(np.clip(z_index + dz, 0, z_max - 1))
        channels.extend([flair[..., z], t1ce[..., z], t2[..., z]])
    return np.stack(channels, axis=-1)

def dice_score_gpu(pred, target, thr=0.5, smooth=1e-5):
    p = (torch.sigmoid(pred) > thr).float()
    intersection = (p * target).sum((1, 2, 3))
    union = p.sum((1, 2, 3)) + target.sum((1, 2, 3))
    return torch.where(union == 0, torch.ones_like(intersection),
                       (2. * intersection + smooth) / (union + smooth))

def precision_score_gpu(pred, target, thr=0.5, smooth=1e-5):
    p = (torch.sigmoid(pred) > thr).float()
    true_pos  = (p * target).sum((1, 2, 3))
    false_pos = (p * (1 - target)).sum((1, 2, 3))
    return (true_pos + smooth) / (true_pos + false_pos + smooth)

def recall_score_gpu(pred, target, thr=0.5, smooth=1e-5):
    p = (torch.sigmoid(pred) > thr).float()
    true_pos = (p * target).sum((1, 2, 3))
    false_neg = ((1 - p) * target).sum((1, 2, 3))
    return (true_pos + smooth) / (true_pos + false_neg + smooth)


def clean_binary_mask(mask, min_size):
    """Removes small, noisy components from a binary mask using connected components."""
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=8)
    cleaned_mask = np.zeros_like(mask)
    for i in range(1, num_labels): # Skip background label 0
        if stats[i, cv2.CC_STAT_AREA] >= min_size:
            cleaned_mask[labels == i] = 1
    return cleaned_mask



def files_from_patient_ids(patient_ids, processed_dir):
    """Gets all slice file paths corresponding to a list of patient IDs."""
    patient_suffixes = {pid.split('_')[-1] for pid in patient_ids}
    all_files = os.listdir(os.path.join(processed_dir, 'images'))
    return sorted([f for f in all_files if f.split('_')[1] in patient_suffixes])

class Brats2_5D_Dataset(Dataset):
    def __init__(self, file_list, base_dir, return_fname=False):
        self.file_list, self.base_dir, self.return_fname = file_list, base_dir, return_fname
    def __len__(self): return len(self.file_list)
    def __getitem__(self, idx):
        fname = self.file_list[idx]
        image = torch.load(os.path.join(self.base_dir, 'images', fname))
        mask_path = os.path.join(self.base_dir, 'masks', fname)
        mask = torch.load(mask_path) if os.path.exists(mask_path) else torch.zeros(1, Config.IMG_SIZE, Config.IMG_SIZE)
        return (image, mask, fname) if self.return_fname else (image, mask)

class CombinedDataset(Dataset):
    """Dataset for the student model, loading from real masks or pseudo-masks."""
    def __init__(self, file_list, config, labeled_patient_suffixes):
        self.file_list, self.config = file_list, config
        self.labeled_suffixes = labeled_patient_suffixes
    def __len__(self): return len(self.file_list)
    def __getitem__(self, idx):
        fname = self.file_list[idx]
        image = torch.load(os.path.join(self.config.PROCESSED_DIR, 'images', fname))
        # Determine if the slice is from a labeled patient or needs a pseudo-label
        is_labeled = fname.split('_')[1] in self.labeled_suffixes
        mask_path = os.path.join(self.config.PROCESSED_DIR, 'masks' if is_labeled else self.config.PSEUDO_LABEL_DIR, fname)
        mask = torch.load(mask_path) if os.path.exists(mask_path) else torch.zeros(1, self.config.IMG_SIZE, self.config.IMG_SIZE)
        return image, mask

def create_balanced_sampler(files, neg_ratio=0.25):
    """Creates a WeightedRandomSampler to balance positive and negative slices."""
    pos_indices = [i for i, f in enumerate(files) if '_pos.pt' in f]
    neg_indices = [i for i, f in enumerate(files) if '_neg.pt' in f]
    if not pos_indices or not neg_indices: return None
    weights = torch.zeros(len(files))
    weights[pos_indices] = (1.0 - neg_ratio) / max(1, len(pos_indices))
    weights[neg_indices] = neg_ratio / max(1, len(neg_indices))
    return torch.utils.data.WeightedRandomSampler(weights, num_samples=len(weights), replacement=True)

def create_sampler_for_student_fast(files, pseudo_dir, labeled_patient_suffixes, neg_ratio=0.25):
    """
    Faster sampler: classifies positives/negatives using positives.json (written during pseudo-generation).
    - For labeled slices: uses filename tag ('_pos.pt' vs '_neg.pt').
    - For pseudo slices: uses positives.json in pseudo_dir.
    """
    pos_indices, neg_indices = [], []

    # load positives.json if exists
    positives_path = os.path.join(pseudo_dir, "positives.json")
    pseudo_map = {}
    if os.path.exists(positives_path):
        with open(positives_path, "r") as f:
            pseudo_map = json.load(f)

    for i, f in enumerate(files):
        is_labeled = f.split('_')[1] in labeled_patient_suffixes
        if is_labeled:
            is_positive = ('_pos.pt' in f)
        else:
            # if not in map (uncertain slice skipped during pseudo gen), treat as negative for sampling
            is_positive = bool(pseudo_map.get(f, False))
        (pos_indices if is_positive else neg_indices).append(i)

    if not pos_indices or not neg_indices:
        return None

    weights = torch.zeros(len(files))
    weights[pos_indices] = (1.0 - neg_ratio) / max(1, len(pos_indices))
    weights[neg_indices] = neg_ratio / max(1, len(neg_indices))
    return torch.utils.data.WeightedRandomSampler(weights, num_samples=len(weights), replacement=True)


class UNet(nn.Module):
    """A standard U-Net model with GroupNorm and Dropout."""
    def __init__(self, in_channels, out_channels, p_drop):
        super().__init__()
        def double_conv(i, o):
            return nn.Sequential(
                nn.Conv2d(i, o, 3, padding=1, bias=False), nn.GroupNorm(8, o), nn.ReLU(inplace=True),
                nn.Conv2d(o, o, 3, padding=1, bias=False), nn.GroupNorm(8, o), nn.ReLU(inplace=True)
            )
        self.dconv_down1 = double_conv(in_channels, 64)
        self.dconv_down2 = double_conv(64, 128)
        self.dconv_down3 = double_conv(128, 256)
        self.dconv_down4 = double_conv(256, 512)
        self.dropout = nn.Dropout2d(p=p_drop)
        self.pool = nn.MaxPool2d(2)
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.dconv_up3 = double_conv(256 + 512, 256)
        self.dconv_up2 = double_conv(128 + 256, 128)
        self.dconv_up1 = double_conv(64 + 128, 64)
        self.conv_last = nn.Conv2d(64, out_channels, 1)

    def forward(self, x):
        c1 = self.dconv_down1(x); p1 = self.pool(c1)
        c2 = self.dconv_down2(p1); p2 = self.pool(c2)
        c3 = self.dconv_down3(p2); p3 = self.pool(c3)
        c4 = self.dropout(self.dconv_down4(p3))
        u3 = self.upsample(c4);   c5 = self.dconv_up3(torch.cat([u3, c3], 1))
        u2 = self.upsample(c5);   c6 = self.dconv_up2(torch.cat([u2, c2], 1))
        u1 = self.upsample(c6);   c7 = self.dconv_up1(torch.cat([u1, c1], 1))
        return self.conv_last(c7)

class TeacherLoss(nn.Module):
    """Combined loss for the teacher model (BCE, Dice, Hausdorff)."""
    def __init__(self, bce_w, dice_w, hausdorff_w, smooth=1e-5):
        super().__init__(); self.bce_w,self.dice_w,self.hausdorff_w,self.smooth = bce_w,dice_w,hausdorff_w,smooth
        self.bce = nn.BCEWithLogitsLoss()
        self.hausdorff = HausdorffDTLoss(include_background=False, to_onehot_y=False, sigmoid=True) if hausdorff_w > 0 else None
    def forward(self, pred, target):
        bce = self.bce(pred, target)
        p = torch.sigmoid(pred)
        i = (p*target).sum((1,2,3)); u = p.sum((1,2,3)) + target.sum((1,2,3))
        dice = 1 - ((2.*i + self.smooth) / (u + self.smooth)).mean()
        loss = self.bce_w*bce + self.dice_w*dice

        if self.hausdorff and self.training and target.max() > 0:
            loss_hd = self.hausdorff(pred.float(), target.float())
            loss = loss + self.hausdorff_w * loss_hd
        return loss

class StudentLoss(nn.Module):
    """Combined loss for the student model using soft pseudo-labels with sharpening and bootstrapping."""
    def __init__(self, bce_w, dice_w, hausdorff_w, temp, beta, smooth=1e-5):
        super().__init__(); self.bce_w,self.dice_w,self.hausdorff_w,self.temp,self.beta,self.smooth = bce_w,dice_w,hausdorff_w,temp,beta,smooth
        self.hausdorff = HausdorffDTLoss(include_background=False, to_onehot_y=False, sigmoid=True) if hausdorff_w > 0 else None
    def sharpen(self, p):
        p_sharp=torch.pow(p, 1./self.temp)
        return p_sharp / (p_sharp + torch.pow(1.-p, 1./self.temp) + 1e-8)
    def forward(self, pred, target):
        p_s = torch.sigmoid(pred)
        with torch.no_grad():
            p_t_sharp = self.sharpen(target)
            mixed_target = self.beta * p_t_sharp + (1. - self.beta) * p_s.detach()

        bce_loss = nn.functional.binary_cross_entropy_with_logits(pred, mixed_target)
        inter = (p_s * p_t_sharp).sum((1,2,3))
        union = p_s.sum((1,2,3)) + p_t_sharp.sum((1,2,3))
        dice_loss = 1 - ((2.*inter + self.smooth) / (union + self.smooth)).mean()

        loss = self.bce_w*bce_loss + self.dice_w*dice_loss
        if self.hausdorff and self.training and target.max() > 0:
            bin_target = (p_t_sharp > 0.5).float()
            loss_hd = self.hausdorff(pred.float(), bin_target.float())
            loss = loss + self.hausdorff_w * loss_hd

        return loss

def train_one_epoch(loader, model, optimizer, loss_fn, scaler, device, grad_clip_norm, amp_enabled=False):
    model.train(); loss_fn.train(); running_loss = 0.0
    progress_bar = tqdm(loader, desc="Training", leave=False)
    devtype = 'cuda' if 'cuda' in str(device) else 'cpu'
    use_amp = (devtype == 'cuda') and bool(amp_enabled)

    for images, masks in progress_bar:
        images, masks = images.to(device), masks.to(device)
        optimizer.zero_grad(set_to_none=True)
        with torch.amp.autocast(device_type=devtype, enabled=use_amp):
            preds = model(images)
            loss = loss_fn(preds, masks)
        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=grad_clip_norm)
        scaler.step(optimizer)
        scaler.update()
        running_loss += loss.item()
        progress_bar.set_postfix(loss=f"{loss.item():.4f}")

    return running_loss / max(1, len(loader))


def evaluate_model(loader, model, loss_fn, device, thr=None, thr_list=(0.30, 0.40, 0.50, 0.60, 0.70), optimize='dice', amp_enabled=False):
    """
    Evaluate model and optionally sweep thresholds on the given loader.
    If `thr` is provided -> evaluate at that threshold only.
    Else -> sweep over `thr_list` and pick the one that optimizes `optimize` ('hd95' or 'dice').
    Returns a dict of averaged metrics at the chosen threshold and includes avg['thr'].
    """
    model.eval(); loss_fn.eval()

    thrs = [thr] if thr is not None else list(thr_list)
    stat = {
        t: {
            'loss': 0.0, 'dice_sum': 0.0, 'prec_sum': 0.0, 'rec_sum': 0.0,
            'neg_acc_sum': 0.0, 'hd95_sum': 0.0,
            'cnt_batches': 0, 'cnt_pos': 0, 'cnt_neg': 0, 'cnt_hd_batches': 0
        } for t in thrs
    }
    hd_metrics = {
        t: HausdorffDistanceMetric(include_background=False, percentile=95, reduction="mean_batch")
        for t in thrs
    }

    def dice_prec_rec(pred_bin, target, smooth=1e-5):
        inter = (pred_bin * target).sum((1,2,3))
        pred_sum = pred_bin.sum((1,2,3))
        tgt_sum  = target.sum((1,2,3))
        dice = (2*inter + smooth) / (pred_sum + tgt_sum + smooth)
        prec = (inter + smooth) / (pred_sum + smooth)
        rec  = (inter + smooth) / (tgt_sum + smooth)
        return dice, prec, rec

    devtype = 'cuda' if ('cuda' in str(device)) else 'cpu'
    use_amp = (devtype == 'cuda') and bool(amp_enabled)

    with torch.inference_mode():
        for images, masks in loader:
            # simple tensor moves (no non_blocking, no channels_last)
            images = images.to(device)
            masks  = masks.to(device)

            with torch.amp.autocast(device_type=devtype, enabled=use_amp):
                logits = model(images)
                batch_loss = loss_fn(logits, masks).item()

            probs_cpu = torch.sigmoid(logits).cpu()
            masks_cpu = masks.cpu()

            is_pos = masks_cpu.sum((1,2,3)) > 0
            any_pos = bool(is_pos.any())
            any_neg = bool((~is_pos).any())

            for t in thrs:
                s = stat[t]
                s['loss'] += batch_loss
                s['cnt_batches'] += 1
                pred_bin = (probs_cpu > t).float()

                if any_pos:
                    pos_idx = is_pos.nonzero(as_tuple=False).squeeze(1)
                    dice, prec, rec = dice_prec_rec(pred_bin[pos_idx], masks_cpu[pos_idx])
                    s['dice_sum'] += dice.sum().item()
                    s['prec_sum'] += prec.sum().item()
                    s['rec_sum']  += rec.sum().item()
                    s['cnt_pos']  += int(pos_idx.numel())

                    hd_val = hd_metrics[t](y_pred=pred_bin[pos_idx], y=masks_cpu[pos_idx])
                    if torch.isfinite(hd_val).all():
                        s['hd95_sum'] += float(hd_val.mean().item())
                        s['cnt_hd_batches'] += 1
                    hd_metrics[t].reset()

                if any_neg:
                    neg_idx = (~is_pos).nonzero(as_tuple=False).squeeze(1)
                    s['neg_acc_sum'] += (pred_bin[neg_idx].sum((1,2,3)) == 0).float().sum().item()
                    s['cnt_neg'] += int(neg_idx.numel())

    def finalize(s):
        return {
            'loss':      s['loss'] / max(1, s['cnt_batches']),
            'dice':      s['dice_sum'] / max(1, s['cnt_pos']),
            'precision': s['prec_sum'] / max(1, s['cnt_pos']),
            'recall':    s['rec_sum']  / max(1, s['cnt_pos']),
            'neg_acc':   s['neg_acc_sum'] / max(1, s['cnt_neg']),
            'hd95':      s['hd95_sum'] / max(1, s['cnt_hd_batches']),
        }

    results = {t: finalize(stat[t]) for t in thrs}

    if thr is not None:
        best_thr = thr
    else:
        if optimize == 'hd95':
            best_thr = min(thrs, key=lambda t: (results[t]['hd95'], -results[t]['dice']))
        else:
            best_thr = max(thrs, key=lambda t: (results[t]['dice'], -results[t]['hd95']))

    out = results[best_thr]
    out['thr'] = float(best_thr)
    return out

def plot_history(history, stage_name, save_dir=None,
                 loss_axis_mode="zero",
                 loss_min_range=0.08,
                 loss_pad_frac=0.10,
                 clamp_floor_to_zero=True,
                 dpi=500,
                 show=True):
    if save_dir:
        os.makedirs(save_dir, exist_ok=True)

    def _compute_ylim(series, mode):
        vals = np.array(series, dtype=float)
        lo, hi = float(vals.min()), float(vals.max())
    
        if mode == "zero":
            lo = 0.0
            span = max(hi - lo, loss_min_range)
            hi = lo + span
        else:
            span = hi - lo
            if span < loss_min_range:
                mid = 0.5 * (hi + lo)
                lo = mid - 0.5 * loss_min_range
                hi = mid + 0.5 * loss_min_range
    
        # pad (used for the Loss panel)
        pad = loss_pad_frac * max(hi - lo, loss_min_range)
        lo -= pad
        if clamp_floor_to_zero:
            lo = max(0.0, lo)
        hi += pad
    
        return lo, hi

    plots = [
        ('Loss', 'Loss', ['train_loss', 'val_loss']),
        ('Val Dice', 'Dice', ['val_dice']),
        ('Val HD95', 'HD95', ['val_hd95']),
        ('Val Precision', 'Precision', ['val_precision']),
        ('Val Recall', 'Recall', ['val_recall'])
    ]
    colors = plt.cm.viridis(np.linspace(0, 1, 7))

    n_plots = len(plots)
    n_cols = 2
    n_rows = int(np.ceil(n_plots / n_cols))

    for idx, (title, ylab, keys) in enumerate(plots):
        fig, ax = plt.subplots(1, 1, figsize=(10, 6))
        all_series = []
        for j, key in enumerate(keys):
            if key in history and history[key]:
                ax.plot(history[key],
                        label=key.replace('_', ' ').title(),
                        marker='o', markersize=4,
                        color=colors[j if len(keys) > 1 else idx+1])
                all_series.extend(history[key])

        if all_series:
            if title == "Loss":
                lo, hi = _compute_ylim(all_series, loss_axis_mode)
                ax.set_ylim(lo, hi)
            else:
                lo, hi = min(all_series), max(all_series)
                if hi - lo < 0.02:
                    mid = 0.5 * (hi + lo)
                    lo, hi = mid - 0.01, mid + 0.01
                pad_abs = (hi - lo) * 0.07
                ax.set_ylim(lo - pad_abs, hi + pad_abs)

        ax.set(title=title, xlabel='Epoch', ylabel=ylab)
        ax.grid(True, ls='--', alpha=0.6)
        ax.legend()
        fig.suptitle(f"{stage_name} - {title}", fontsize=16)

        if save_dir:
            out_path = os.path.join(save_dir, f"{stage_name}_{title.replace(' ', '_')}.png")
            fig.savefig(out_path, dpi=dpi, bbox_inches='tight')
        if show:
            plt.show()
        plt.close(fig)

def generate_pseudo_masks(model, loader, config, output_dir):
    """
    Use the teacher model to generate soft pseudo-labels.
    Writes 'positives.json' summarizing which slices were positive.
    """
    model.eval()
    os.makedirs(output_dir, exist_ok=True)
    saved_count = 0
    positive_map = {}

    # device type for autocast
    if isinstance(config.DEVICE, torch.device):
        devtype = config.DEVICE.type
    else:
        devtype = 'cuda' if 'cuda' in str(config.DEVICE).lower() else 'cpu'

    for images, _, fnames in tqdm(loader, desc="Generating Pseudo-Masks", leave=False):
        # simple move to device (no non_blocking, no channels_last)
        images = images.to(config.DEVICE)

        with torch.inference_mode(), torch.amp.autocast(device_type=devtype, enabled=config.AMP_ENABLED):
            logits = model(images)
            probs = torch.sigmoid(logits).cpu()  # move once to CPU to free GPU mem

        for i, fname in enumerate(fnames):
            # probs[i]: (1, H, W) on CPU
            prob_np = probs[i].squeeze(0).numpy()  # -> (H, W)
            binary_mask = (prob_np > config.PSEUDO_BINARIZE_THR).astype(np.uint8)
            cleaned = clean_binary_mask(binary_mask, config.PSEUDO_MIN_SIZE)

            if cleaned.sum() > 0:
                # positive pseudo: mask the soft probs with cleaned components
                soft_pseudo = probs[i] * torch.from_numpy(cleaned).unsqueeze(0).float()
                torch.save(soft_pseudo, os.path.join(output_dir, fname))
                saved_count += 1
                positive_map[fname] = True
            elif prob_np.mean() < config.CONFIDENT_NEGATIVE_THR:
                # confident negative
                torch.save(torch.zeros_like(probs[i]), os.path.join(output_dir, fname))
                saved_count += 1
                positive_map[fname] = False
            # else: uncertain -> skip writing

    with open(os.path.join(output_dir, "positives.json"), "w") as f:
        json.dump(positive_map, f)

    return saved_count
    

def save_final_visuals_grid(
    pred_model,
    teacher_model_for_pseudo,
    loader_with_fnames,
    device,
    config,
    num_examples=10,
    save_path=None,
    thr=0.5,
    show=True
):
    """
    For up to `num_examples` DIFFERENT patients:
      • choose the slice with the LARGEST GT tumor area per patient
      • draw 4 columns: [FLAIR, GT, Pseudo, Pred]
      • save 10 individual 1x4 rows + one combined 10x4 image (higher DPI)

    Notes
    -----
    - Patient ID is inferred from the filename format used in preprocessing:
      "slice_{patientSuffix}_{z:03d}_{counter}_{tag}.pt"
      -> patient id = tokens[1]
    - If `save_path` is a directory, files go inside it. If it's a filename,
      its directory is used and that file path is used for the combined image.
    """
    import os, re
    import numpy as np
    import torch
    import matplotlib.pyplot as plt
    from tqdm import tqdm

    pred_model.eval()
    if teacher_model_for_pseudo is not None:
        teacher_model_for_pseudo.eval()

    ds = loader_with_fnames.dataset
    devtype = 'cuda' if getattr(device, "type", str(device)).startswith('cuda') else 'cpu'

    def patient_id_from_fname(fname: str) -> str:
        toks = str(fname).split('_')
        return toks[1] if len(toks) > 1 else str(fname)

    def to_numpy_grayscale(t):
        if t.ndim == 3 and t.shape[0] == 1:
            t = t[0]
        return t.detach().cpu().numpy()

    def predict_bin(image_):
        with torch.no_grad(), torch.amp.autocast(device_type=devtype, enabled=bool(config.AMP_ENABLED)):
            logits = pred_model(image_.unsqueeze(0).to(device))
            prob_pred = torch.sigmoid(logits).squeeze(0)
        return (prob_pred > thr).float(), prob_pred

    def pseudo_bin(image_, fname_):
        pseudo_path = None
        if hasattr(config, "PSEUDO_LABEL_DIR") and config.PSEUDO_LABEL_DIR:
            pseudo_path = os.path.join(config.PSEUDO_LABEL_DIR, fname_)
        if pseudo_path and os.path.exists(pseudo_path):
            pseudo_tensor = torch.load(pseudo_path, map_location='cpu')
            return (pseudo_tensor > thr).float()

        if teacher_model_for_pseudo is not None:
            with torch.no_grad(), torch.amp.autocast(device_type=devtype, enabled=bool(config.AMP_ENABLED)):
                t_logits = teacher_model_for_pseudo(image_.unsqueeze(0).to(device))
                t_prob = torch.sigmoid(t_logits).squeeze(0)
            t_np = to_numpy_grayscale(t_prob)
            t_bin = (t_np > getattr(config, "PSEUDO_BINARIZE_THR", thr)).astype(np.uint8)
            if 'clean_binary_mask' in globals():
                t_bin = clean_binary_mask(t_bin, getattr(config, "PSEUDO_MIN_SIZE", 0))
            return torch.from_numpy(t_bin).unsqueeze(0).float()

        return None

    best_idx_per_patient = {}
    for i in range(len(ds)):
        _, gt_mask, fname = ds[i]
        area = float(gt_mask.sum().item())
        if area <= 0:
            continue
        pid = patient_id_from_fname(fname)
        if (pid not in best_idx_per_patient) or (area > best_idx_per_patient[pid][0]):
            best_idx_per_patient[pid] = (area, i)

    if not best_idx_per_patient:
        print("No positive slices found; cannot build the requested grid.")
        return

    sorted_best = sorted(best_idx_per_patient.items(), key=lambda kv: -kv[1][0])
    chosen = sorted_best[:num_examples]
    chosen_indices = [v[1][1] for v in chosen]
    chosen_pids    = [k for k, _ in chosen]

    if save_path:
        if os.path.splitext(save_path)[1]:
            save_dir = os.path.dirname(save_path) or "."
            combined_path = save_path
        else:
            save_dir = save_path
            combined_path = os.path.join(save_dir, "final_grid.png")
    else:
        save_dir = os.path.join(getattr(config, "OUTPUT_DIR", "."), "final_visuals")
        combined_path = os.path.join(save_dir, "final_grid.png")

    os.makedirs(save_dir, exist_ok=True)
    row_paths = []

    shown_warning_once = False
    for ri, (idx, pid) in enumerate(tqdm(zip(chosen_indices, chosen_pids),
                                         total=len(chosen_indices),
                                         desc="Saving 1x4 rows",
                                         leave=False)):
        image, gt_mask, fname = ds[idx]

        flair_c = image[3].to(device).unsqueeze(0)
        pred_bin, _ = predict_bin(image)
        pseudo = pseudo_bin(image, fname)
        if pseudo is None:
            if not shown_warning_once:
                print("Warning: No pseudo-label for some slices; pseudo column will be blank.")
                shown_warning_once = True
            pseudo = torch.zeros_like(pred_bin)

        fig_row, ax = plt.subplots(1, 4, figsize=(16,4))
        ax[0].imshow(to_numpy_grayscale(flair_c), cmap='gray', vmin=0, vmax=1); ax[0].set_title("FLAIR");  ax[0].axis('off')
        ax[1].imshow(to_numpy_grayscale(gt_mask), cmap='gray', vmin=0, vmax=1); ax[1].set_title("GT Mask"); ax[1].axis('off')
        ax[2].imshow(to_numpy_grayscale(pseudo),  cmap='gray', vmin=0, vmax=1); ax[2].set_title("Pseudo");  ax[2].axis('off')
        ax[3].imshow(to_numpy_grayscale(pred_bin), cmap='gray', vmin=0, vmax=1); ax[3].set_title("Pred");    ax[3].axis('off')
        plt.tight_layout()
        fig_row.subplots_adjust(wspace=0.05)

        row_path = os.path.join(save_dir, f"row_{ri+1:02d}_patient-{pid}.png")
        fig_row.savefig(row_path, dpi=500, bbox_inches='tight')
        if show:
            plt.show()
        plt.close(fig_row)
        row_paths.append(row_path)

    nrows, ncols = len(chosen_indices), 4
    fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * 4, nrows * 4))
    if nrows == 1:
        axes = np.expand_dims(axes, axis=0)

    for ri, (idx, pid) in enumerate(zip(chosen_indices, chosen_pids)):
        image, gt_mask, fname = ds[idx]
        flair_c = image[3].to(device).unsqueeze(0)
        pred_bin, _ = predict_bin(image)
        pseudo = pseudo_bin(image, fname)
        if pseudo is None:
            pseudo = torch.zeros_like(pred_bin)

        axes[ri, 0].imshow(to_numpy_grayscale(flair_c), cmap='gray', vmin=0, vmax=1); axes[ri, 0].set_title("FLAIR");  axes[ri, 0].axis('off')
        axes[ri, 1].imshow(to_numpy_grayscale(gt_mask), cmap='gray', vmin=0, vmax=1); axes[ri, 1].set_title("GT Mask"); axes[ri, 1].axis('off')
        axes[ri, 2].imshow(to_numpy_grayscale(pseudo),  cmap='gray', vmin=0, vmax=1); axes[ri, 2].set_title("Pseudo");  axes[ri, 2].axis('off')
        axes[ri, 3].imshow(to_numpy_grayscale(pred_bin), cmap='gray', vmin=0, vmax=1); axes[ri, 3].set_title("Pred");    axes[ri, 3].axis('off')

    plt.tight_layout()
    fig.subplots_adjust(wspace=0.05)
    fig.savefig(combined_path, dpi=500, bbox_inches='tight')
    if show:
        plt.show()
    plt.close(fig)

    print(f"Saved combined grid: {combined_path}")
    for p in row_paths:
        print(f"Saved row: {p}")

    return {"combined": combined_path, "rows": row_paths}




### Pre-computation Step 1: Data Preprocessing


def run_preprocessing(config):
    """Converts the raw 3D NIfTI files into 2.5D (9-channel) PyTorch tensors."""

    print("\nStarting 2.5D Preprocessing (9-channel with Hard Negative Mining)")
    for subdir in ['images', 'masks']:
        os.makedirs(os.path.join(config.PROCESSED_DIR, subdir), exist_ok=True)

    all_patient_dirs = sorted(
        [d for d in glob.glob(os.path.join(config.BASE_DIR, 'BraTS20_Training_*')) if os.path.isdir(d)]
    )

    def _save_slice(img_stack, mask, patient_id, z, counter, tag):
        img_crop, bbox = crop_to_brain(img_stack)
        if not bbox:
            return 0
    
        # dtypes
        img_crop = img_crop.astype(np.float32, copy=False)
        mask_crop = mask[bbox[0]:bbox[2], bbox[1]:bbox[3]].astype(np.uint8, copy=False)
    
        # per-channel normalize
        for i in range(img_crop.shape[2]):
            chan = img_crop[:, :, i]
            cmin, cmax = chan.min(), chan.max()
            if cmax > cmin:
                img_crop[:, :, i] = (chan - cmin) / (cmax - cmin + 1e-6)
            else:
                img_crop[:, :, i] = 0.0
    
        h, w, _ = img_crop.shape
        scale = config.IMG_SIZE / max(h, w)
        nw, nh = int(w * scale), int(h * scale)
    
        resized_chs = []
        for i in range(img_crop.shape[2]):
            resized_chs.append(cv2.resize(img_crop[:, :, i], (nw, nh), interpolation=cv2.INTER_AREA))
        img_rz = np.stack(resized_chs, axis=2)  # (nh, nw, C)
    
        mask_rz = cv2.resize(mask_crop, (nw, nh), interpolation=cv2.INTER_NEAREST)
    
        # --- pad with numpy (works for any #channels) ---
        dw, dh = config.IMG_SIZE - nw, config.IMG_SIZE - nh
        top, left = dh // 2, dw // 2
    
        final_img = np.pad(
            img_rz,
            pad_width=((top, dh - top), (left, dw - left), (0, 0)),
            mode="constant",
            constant_values=0,
        )
        final_mask = np.pad(
            mask_rz,
            pad_width=((top, dh - top), (left, dw - left)),
            mode="constant",
            constant_values=0,
        )
    
        fname = f"slice_{patient_id.split('_')[-1]}_{z:03d}_{counter}_{tag}.pt"
        torch.save(torch.from_numpy(final_img).permute(2, 0, 1).float(),
                   os.path.join(config.PROCESSED_DIR, 'images', fname))
        torch.save(torch.from_numpy(final_mask).unsqueeze(0).float(),
                   os.path.join(config.PROCESSED_DIR, 'masks', fname))
        return 1

    total_saved = 0
    for patient_dir in tqdm(all_patient_dirs, desc="Preprocessing Patients"):
        pid = os.path.basename(patient_dir)
        try:
            vols = {
                'flair': nib.load(get_file_path(config.BASE_DIR, pid, 'flair')).get_fdata(dtype=np.float32),
                't1ce' : nib.load(get_file_path(config.BASE_DIR, pid, 't1ce' )).get_fdata(dtype=np.float32),
                't2'   : nib.load(get_file_path(config.BASE_DIR, pid, 't2'   )).get_fdata(dtype=np.float32),
                'seg'  : nib.load(get_file_path(config.BASE_DIR, pid, 'seg'  )).get_fdata(dtype=np.float32),
            }
            vols['seg'] = (vols['seg'] > 0).astype(np.uint8)
        except Exception as e:
            print(f"Skipping {pid}: {e}")
            continue

        pos_indices = [z for z in range(vols['seg'].shape[2]) if vols['seg'][:, :, z].sum() > 0]
        if not pos_indices:
            del vols
            gc.collect()
            continue

        # Save positive slices
        for z in pos_indices:
            img_9ch = stack_neighbors(vols['flair'], vols['t1ce'], vols['t2'], z)
            total_saved += _save_slice(img_9ch, vols['seg'][:, :, z], pid, z, total_saved, "pos")

        # Mine for hard negative slices near the tumor region
        z_min, z_max = min(pos_indices), max(pos_indices)
        neg_cand = [
            z for z in range(max(0, z_min - 5), min(vols['seg'].shape[2], z_max + 6))
            if vols['seg'][:, :, z].sum() == 0
        ]
        num_to_sample = min(config.HARD_NEGATIVE_COUNT, len(neg_cand))
        if num_to_sample > 0:
            for z in random.sample(neg_cand, k=num_to_sample):
                img_9ch = stack_neighbors(vols['flair'], vols['t1ce'], vols['t2'], z)
                total_saved += _save_slice(img_9ch, np.zeros_like(vols['seg'][:, :, z], dtype=np.uint8),
                                           pid, z, total_saved, "neg")

        # Free memory for this patient
        del vols
        gc.collect()

    print(f" Preprocessing Complete :- Saved {total_saved} total 2.5D slices.")

if not os.path.exists(Config.PROCESSED_DIR) or not os.listdir(os.path.join(Config.PROCESSED_DIR, 'images')):
    run_preprocessing(Config)
else:
    print(f"Preprocessed data found in '{Config.PROCESSED_DIR}'. Skipping preprocessing.")




### Part 2.5: Visualize Preprocessing Pipeline



print("\n Visualizing the Preprocessing Pipeline: Original vs. Processed ")

def _preprocess_for_viz(original_slice, target_size):
    cropped_slice, _ = crop_to_brain(original_slice, optional_padding=0)
    if cropped_slice.shape[0] == 0 or cropped_slice.shape[1] == 0:
        return np.zeros((target_size, target_size)), np.zeros_like(original_slice)
    h, w = cropped_slice.shape
    scale = target_size / max(h, w)
    new_w, new_h = int(w * scale), int(h * scale)
    resized = cv2.resize(cropped_slice, (new_w, new_h), interpolation=cv2.INTER_AREA)
    delta_w, delta_h = target_size - new_w, target_size - new_h
    top, left = delta_h // 2, delta_w // 2
    final_img = cv2.copyMakeBorder(resized, top, delta_h - top, left, delta_w - left, cv2.BORDER_CONSTANT, value=0)
    return final_img, cropped_slice

if all_patient_dirs_vis:
    sample_patient_id_2 = os.path.basename(all_patient_dirs_vis[20])  # keep your selection
    try:
        flair_vol_vis2 = nib.load(get_vis_file_path(Config.BASE_DIR, sample_patient_id_2, 'flair')).get_fdata()
        fig, axes = plt.subplots(3, 4, figsize=(16, 12))
        plt.suptitle("Preprocessing Pipeline: \n Original -> Cropped -> Resized", fontsize=18)

        row_images = []  # store row outputs for later

        for i in range(3):
            slice_idx = flair_vol_vis2.shape[2] // 2 - 15 + i*15
            original_slice = flair_vol_vis2[:, :, slice_idx]
            processed_128, cropped_slice = _preprocess_for_viz(original_slice, 128)
            processed_64, _              = _preprocess_for_viz(original_slice, 64)

            # --- Draw in the big grid ---
            axes[i, 0].set_ylabel(f"Sample Slice {i+1}", fontsize=14, rotation=90, labelpad=20)
            axes[i, 0].imshow(original_slice, cmap='bone'); axes[i, 0].set_title(f"Original\nSize: {original_slice.shape}"); axes[i, 0].axis('off')
            axes[i, 1].imshow(cropped_slice, cmap='bone'); axes[i, 1].set_title(f"1. Brain Cropped\nSize: {cropped_slice.shape}"); axes[i, 1].axis('off')
            axes[i, 2].imshow(processed_128, cmap='bone'); axes[i, 2].set_title(f"2. Padded & Resized\nSize: (128, 128)"); axes[i, 2].axis('off')
            axes[i, 3].imshow(processed_64, cmap='bone'); axes[i, 3].set_title(f"Comparison: (64, 64)\n(Shows Detail Loss)"); axes[i, 3].axis('off')

            # --- Also make a separate 1×4 figure for this row ---
            fig_row, row_axes = plt.subplots(1, 4, figsize=(16, 4))
            row_axes[0].imshow(original_slice, cmap='bone'); row_axes[0].set_title(f"Original\n{original_slice.shape}"); row_axes[0].axis('off')
            row_axes[1].imshow(cropped_slice, cmap='bone'); row_axes[1].set_title(f"1. Brain Cropped\n{cropped_slice.shape}"); row_axes[1].axis('off')
            row_axes[2].imshow(processed_128, cmap='bone'); row_axes[2].set_title("2. Padded & Resized\n(128, 128)"); row_axes[2].axis('off')
            row_axes[3].imshow(processed_64, cmap='bone'); row_axes[3].set_title("Comparison (64, 64)\nDetail Loss"); row_axes[3].axis('off')
            plt.tight_layout()

            row_out = os.path.join(Config.OUTPUT_DIR, f"preprocessing_row{i+1}_{sample_patient_id_2}.png")
            fig_row.savefig(row_out, dpi=500, bbox_inches='tight')
            plt.close(fig_row)
            row_images.append(row_out)

        # --- Save the big combined figure ---
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        prep_out = os.path.join(Config.OUTPUT_DIR, f"preprocessing_pipeline_{sample_patient_id_2}.png")
        fig.savefig(prep_out, dpi=500, bbox_inches='tight')
        plt.close(fig)

        print(f"Saved combined pipeline: {prep_out}")
        for p in row_images:
            print(f"Saved row: {p}")

    except Exception as e:
        print(f"Could not generate preprocessing visualization for patient {sample_patient_id_2}: {e}")
        

### Pre-computation Step 2: Create Data Splits

def run_split(config):
    print(" Creating 70/15/15 data split of train, validation and test")

    all_pids = sorted(
        os.path.basename(p)
        for p in glob.glob(os.path.join(config.BASE_DIR, "BraTS20_Training_*"))
        if os.path.isdir(p)
    )
    if not all_pids:
        print("Error: No patient directories found.")
        return

    random.shuffle(all_pids)
    n = len(all_pids)

    n_test = int(round(0.15 * n))
    n_val  = int(round(0.15 * n))

    n_test = min(n_test, max(1, n - 2))
    n_val  = min(n_val,  max(1, n - n_test - 1))

    test_ids = all_pids[:n_test]
    val_ids  = all_pids[n_test:n_test + n_val]
    train_ids = all_pids[n_test + n_val:]

    splits = {
        "train": train_ids,
        "val": val_ids,
        "test": test_ids,
        "meta": {
            "ratios": {"train": 0.70, "val": 0.15, "test": 0.15},
            "seed": getattr(config, "RANDOM_STATE", None),
            "strategy": "patient-level"
        }
    }

    os.makedirs(config.OUTPUT_DIR, exist_ok=True)
    split_path = os.path.join(config.OUTPUT_DIR, "splits.json")
    with open(split_path, "w") as f:
        json.dump(splits, f, indent=2)

    print(f"Saved data splits to {split_path}")
    print(f"Split counts: {len(train_ids)} train, {len(val_ids)} val, {len(test_ids)} test")


split_file_path = os.path.join(Config.OUTPUT_DIR, "splits.json")
if not os.path.exists(split_file_path):
    run_split(Config)
else:
    print(f"Data split file found at '{split_file_path}'. Skipping split.")




### Action 1: Tune Hyperparameters with Optuna




def run_tuning(config):
    """Runs the Optuna hyperparameter tuning study (GPU/AMP if available) and prints full metrics each epoch."""
    print("--- Starting Optuna Hyperparameter Tuning ---")
    split_path = os.path.join(config.OUTPUT_DIR, "splits.json")
    if not os.path.exists(split_path):
        print("Error: `splits.json` not found.")
        return
    with open(split_path, 'r') as f:
        splits = json.load(f)

    # helper for epoch logs
    def _fmt_epoch(stage, ep, ep_total, train_loss, v):
        return (f"{stage} - Ep {ep}/{ep_total} | "
                f"TrainLoss:{train_loss:.4f} | ValLoss:{v['loss']:.4f} | "
                f"Dice:{v['dice']:.4f} | HD95:{v['hd95']:.4f} | "
                f"Prec:{v['precision']:.4f} | Rec:{v['recall']:.4f} | "
                f"NegAcc:{v['neg_acc']:.4f}")

    def objective(trial):
        with tempfile.TemporaryDirectory() as temp_dir:
            cfg = copy.deepcopy(config)

            # ---- device & dataloader tuning ----
            if torch.cuda.is_available():
                cfg.DEVICE = torch.device("cuda:0")
                cfg.AMP_ENABLED = False   
                cfg.NUM_WORKERS = 2
                cfg.PERSISTENT_WORKERS = False
                cfg.PREFETCH_FACTOR = 2

            else:
                cfg.DEVICE = torch.device("cpu")
                cfg.AMP_ENABLED = False
                cfg.NUM_WORKERS = 0
                cfg.PERSISTENT_WORKERS = False
                cfg.PREFETCH_FACTOR = None

            # ---- sample hyperparams ----
            cfg.LEARNING_RATE = trial.suggest_float("lr", 1e-5, 5e-4, log=True)
            cfg.DROPOUT_PROB  = trial.suggest_float("dropout", 0.0, 0.4)
            neg_ratio         = trial.suggest_float("neg_ratio", 0.1, 0.3)
            cfg.PSEUDO_BINARIZE_THR = trial.suggest_float("thr_bin", 0.3, 0.7)
            labeled_ratio     = trial.suggest_float("labeled_ratio", 0.4, 0.6)

            t_bce  = trial.suggest_float("t_bce", 0.1, 1.0)
            t_dice = trial.suggest_float("t_dice", 0.1, 1.0)
            t_hd   = trial.suggest_float("t_hd", 0.1, 1.0)
            s = t_bce + t_dice + t_hd
            cfg.S1_BCE_WEIGHT, cfg.S1_DICE_WEIGHT, cfg.S1_HAUSDORFF_WEIGHT = t_bce/s, t_dice/s, t_hd/s

            s_bce  = trial.suggest_float("s_bce", 0.1, 1.0)
            s_dice = trial.suggest_float("s_dice", 0.1, 1.0)
            s = s_bce + s_dice
            cfg.S3_BCE_WEIGHT, cfg.S3_DICE_WEIGHT = s_bce/s, s_dice/s

            # ---- semi-supervised split ----
            rng = np.random.RandomState(cfg.RANDOM_STATE)
            train_ids = splits['train']
            labeled_ids = sorted(rng.choice(train_ids, size=int(len(train_ids) * labeled_ratio), replace=False).tolist())
            unlabeled_ids = sorted(list(set(train_ids) - set(labeled_ids)))
            labeled_files = files_from_patient_ids(labeled_ids, cfg.PROCESSED_DIR)
            unlabeled_files = files_from_patient_ids(unlabeled_ids, cfg.PROCESSED_DIR)
            val_files = files_from_patient_ids(splits['val'], cfg.PROCESSED_DIR)

            # ---- loaders ----
            train_loader = DataLoader(
                Brats2_5D_Dataset(labeled_files, cfg.PROCESSED_DIR),
                batch_size=cfg.BATCH_SIZE,
                sampler=create_balanced_sampler(labeled_files, neg_ratio),
                **dl_kwargs(cfg),
            )
            val_loader = DataLoader(
                Brats2_5D_Dataset(val_files, cfg.PROCESSED_DIR),
                batch_size=cfg.BATCH_SIZE, shuffle=False, **dl_kwargs(cfg),
            )

            # ---- Stage 1: Teacher ----
            model_t = UNet(9, 1, cfg.DROPOUT_PROB).to(cfg.DEVICE)
            if torch.cuda.device_count() > 1:
                model_t = nn.DataParallel(model_t)
            optim_t = optim.AdamW(model_t.parameters(), lr=cfg.LEARNING_RATE, weight_decay=cfg.WEIGHT_DECAY)
            loss_t  = TeacherLoss(cfg.S1_BCE_WEIGHT, cfg.S1_DICE_WEIGHT, cfg.S1_HAUSDORFF_WEIGHT)
            scaler  = torch.amp.GradScaler(enabled=cfg.AMP_ENABLED)

            for ep in range(cfg.TUNE_EPOCHS_STAGE1):
                train_loss = train_one_epoch(
                    train_loader, model_t, optim_t, loss_t, scaler,
                    cfg.DEVICE, cfg.GRAD_CLIP_NORM, amp_enabled=cfg.AMP_ENABLED
                )
                val_metrics = evaluate_model(
                    val_loader, model_t, loss_t, cfg.DEVICE,
                    amp_enabled=cfg.AMP_ENABLED
                )
                print(_fmt_epoch("S1[Tune]", ep+1, cfg.TUNE_EPOCHS_STAGE1, train_loss, val_metrics))
                trial.report(val_metrics['dice'], ep)
                if trial.should_prune():
                    raise optuna.TrialPruned()

            # ---- Stage 2: pseudo labels ----
            teacher_model = model_t.module if isinstance(model_t, nn.DataParallel) else model_t
            unl_loader = DataLoader(
                Brats2_5D_Dataset(unlabeled_files, cfg.PROCESSED_DIR, return_fname=True),
                batch_size=cfg.BATCH_SIZE, shuffle=False, **dl_kwargs(cfg),
            )
            generate_pseudo_masks(teacher_model, unl_loader, cfg, temp_dir)

            # ---- Stage 3: Student ----
            student_cfg = copy.deepcopy(cfg)
            student_cfg.PSEUDO_LABEL_DIR = temp_dir
            unl_with_pseudo = [f for f in unlabeled_files if os.path.exists(os.path.join(temp_dir, f))]
            combined_files = labeled_files + unl_with_pseudo
            labeled_suffixes = {pid.split('_')[-1] for pid in labeled_ids}

            combined_ds = CombinedDataset(combined_files, student_cfg, labeled_suffixes)
            sampler_s = create_sampler_for_student_fast(
                combined_files, student_cfg.PSEUDO_LABEL_DIR, labeled_suffixes, neg_ratio
            )
            combined_loader = DataLoader(
                combined_ds, batch_size=cfg.BATCH_SIZE, sampler=sampler_s, **dl_kwargs(cfg)
            )

            student = UNet(9, 1, cfg.DROPOUT_PROB).to(cfg.DEVICE)
            student.load_state_dict(teacher_model.state_dict())
            if torch.cuda.device_count() > 1:
                student = nn.DataParallel(student)
            optim_s = optim.AdamW(student.parameters(), lr=cfg.LEARNING_RATE/2, weight_decay=cfg.WEIGHT_DECAY)
            loss_s  = StudentLoss(cfg.S3_BCE_WEIGHT, cfg.S3_DICE_WEIGHT, cfg.S3_HAUSDORFF_WEIGHT, cfg.STUDENT_TEMP, cfg.STUDENT_BOOTSTRAP_BETA)
            scaler_s = torch.amp.GradScaler(enabled=cfg.AMP_ENABLED)

            best_score = -1.0
            for ep in range(cfg.TUNE_EPOCHS_STAGE2):
                train_loss_s3 = train_one_epoch(
                    combined_loader, student, optim_s, loss_s, scaler_s,
                    cfg.DEVICE, cfg.GRAD_CLIP_NORM, amp_enabled=cfg.AMP_ENABLED
                )
                val_metrics   = evaluate_model(
                    val_loader, student, loss_s, cfg.DEVICE,
                    amp_enabled=cfg.AMP_ENABLED
                )
                print(_fmt_epoch("S3[Tune]", ep+1, cfg.TUNE_EPOCHS_STAGE2, train_loss_s3, val_metrics))
                score = val_metrics['dice'] - 0.1 * min(val_metrics['hd95'] / 10.0, 1.0)
                best_score = max(best_score, score)
                trial.report(val_metrics['dice'], cfg.TUNE_EPOCHS_STAGE1 + ep)
                if trial.should_prune():
                    raise optuna.TrialPruned()

            return best_score

    study = optuna.create_study(
        direction="maximize",
        pruner=optuna.pruners.MedianPruner(n_warmup_steps=4)
    )
    study.optimize(objective, n_trials=config.N_TUNING_TRIALS, timeout=3600*8)

    print("\n--- Tuning Complete ---")
    print(f"Best trial score: {study.best_value}")
    print("Best hyperparameters:", study.best_params)
    with open(os.path.join(config.OUTPUT_DIR, "best_params.json"), 'w') as f:
        json.dump(study.best_params, f, indent=2)
    print(f"Saved best parameters to {os.path.join(config.OUTPUT_DIR, 'best_params.json')}")





### Action 2: Final Training and Evaluation




def run_final_training_single(config):
    
    # -------- device & dataloader tuning --------
    cfg = copy.deepcopy(config)
    if torch.cuda.is_available():
        cfg.DEVICE = torch.device("cuda:0")
        cfg.AMP_ENABLED = False   
        cfg.NUM_WORKERS = 2
        cfg.PERSISTENT_WORKERS = False
        cfg.PREFETCH_FACTOR = 2
    else:
        cfg.DEVICE = torch.device("cpu")
        cfg.AMP_ENABLED = False
        cfg.NUM_WORKERS = 0
        cfg.PERSISTENT_WORKERS = False
        cfg.PREFETCH_FACTOR = None

    print(f"[Single] Using device: {cfg.DEVICE}")

    def _fmt_epoch(stage, ep, ep_total, train_loss, v):
        return (f"{stage} - Ep {ep}/{ep_total} | "
                f"TrainLoss:{train_loss:.4f} | ValLoss:{v['loss']:.4f} | "
                f"Dice:{v['dice']:.4f} | HD95:{v['hd95']:.4f} | "
                f"Prec:{v['precision']:.4f} | Rec:{v['recall']:.4f} | "
                f"NegAcc:{v['neg_acc']:.4f}")

    # -------- load splits / best params --------
    split_path  = os.path.join(cfg.OUTPUT_DIR, "splits.json")
    params_path = os.path.join(cfg.OUTPUT_DIR, "best_params.json")
    if not os.path.exists(split_path):
        raise FileNotFoundError("`splits.json` not found.")
    if not os.path.exists(params_path):
        print("Warning: `best_params.json` not found. Using defaults.")
        default_params = {
            "lr": cfg.LEARNING_RATE, "dropout": cfg.DROPOUT_PROB,
            "neg_ratio": 0.25, "thr_bin": cfg.PSEUDO_BINARIZE_THR,
            "t_bce": cfg.S1_BCE_WEIGHT, "t_dice": cfg.S1_DICE_WEIGHT, "t_hd": cfg.S1_HAUSDORFF_WEIGHT,
            "s_bce": cfg.S3_BCE_WEIGHT, "s_dice": cfg.S3_DICE_WEIGHT,
            "labeled_ratio": 0.5,
        }
        with open(params_path, 'w') as f:
            json.dump(default_params, f, indent=2)

    with open(split_path,  'r') as f: splits      = json.load(f)
    with open(params_path, 'r') as f:
        best_params = json.load(f)
    print("--- Using the following hyperparameters for the final run: ---")
    for key, val in best_params.items():
        if isinstance(val, float):
            print(f"- {key:<20}: {val:.5f}")
        else:
            print(f"- {key:<20}: {val}")


    # apply best params
    cfg.LEARNING_RATE       = best_params["lr"]
    cfg.DROPOUT_PROB        = best_params["dropout"]
    cfg.PSEUDO_BINARIZE_THR = best_params["thr_bin"]
    neg_ratio               = best_params["neg_ratio"]

    t_bce, t_dice, t_hd = best_params["t_bce"], best_params["t_dice"], best_params["t_hd"]
    s = t_bce + t_dice + t_hd
    cfg.S1_BCE_WEIGHT, cfg.S1_DICE_WEIGHT, cfg.S1_HAUSDORFF_WEIGHT = t_bce/s, t_dice/s, t_hd/s

    s_bce, s_dice = best_params["s_bce"], best_params["s_dice"]
    s = s_bce + s_dice
    cfg.S3_BCE_WEIGHT, cfg.S3_DICE_WEIGHT = s_bce/s, s_dice/s

    # -------- build labeled/unlabeled/val/test sets --------
    train_ids = splits['train']
    val_ids   = splits['val']
    test_ids  = splits['test']

    labeled_ratio = best_params.get("labeled_ratio", 0.5)
    rng = np.random.RandomState(cfg.RANDOM_STATE)
    labeled_ids   = sorted(rng.choice(train_ids, size=max(1, int(len(train_ids)*labeled_ratio)), replace=False).tolist())
    unlabeled_ids = sorted(list(set(train_ids) - set(labeled_ids)))

    labeled_files_all   = files_from_patient_ids(labeled_ids,   cfg.PROCESSED_DIR)
    unlabeled_files_all = files_from_patient_ids(unlabeled_ids, cfg.PROCESSED_DIR)
    val_files_all       = files_from_patient_ids(val_ids,       cfg.PROCESSED_DIR)
    test_files_all      = files_from_patient_ids(test_ids,      cfg.PROCESSED_DIR)

    # -------- Stage 1: Teacher --------
    print("\n--- Final Run (Single): Stage 1 (Teacher) ---")
    train_loader = DataLoader(
        Brats2_5D_Dataset(labeled_files_all, cfg.PROCESSED_DIR),
        batch_size=cfg.BATCH_SIZE,
        sampler=create_balanced_sampler(labeled_files_all, neg_ratio),
        **dl_kwargs(cfg)
    )
    val_loader = DataLoader(
        Brats2_5D_Dataset(val_files_all, cfg.PROCESSED_DIR),
        batch_size=cfg.BATCH_SIZE, shuffle=False, **dl_kwargs(cfg)
    )

    model_t  = UNet(9, 1, cfg.DROPOUT_PROB).to(cfg.DEVICE)
    optim_t  = optim.AdamW(model_t.parameters(), lr=cfg.LEARNING_RATE, weight_decay=cfg.WEIGHT_DECAY)
    sched_t  = ReduceLROnPlateau(optim_t, 'max', patience=cfg.SCHEDULER_PATIENCE, verbose=True)
    loss_t   = TeacherLoss(cfg.S1_BCE_WEIGHT, cfg.S1_DICE_WEIGHT, cfg.S1_HAUSDORFF_WEIGHT)
    scaler_t = torch.amp.GradScaler(enabled=cfg.AMP_ENABLED)

    history_s1 = {k:[] for k in ['train_loss','val_loss','val_dice','val_hd95','val_neg_acc','val_precision','val_recall']}
    best_dice_s1 = -1.0
    best_epoch_s1 = -1
    best_metrics_s1 = None
    
    for ep in range(cfg.EPOCHS_STAGE1):
        train_loss  = train_one_epoch(
            train_loader, model_t, optim_t, loss_t, scaler_t,
            cfg.DEVICE, cfg.GRAD_CLIP_NORM, amp_enabled=cfg.AMP_ENABLED
        )
        val_metrics = evaluate_model(val_loader, model_t, loss_t, cfg.DEVICE, amp_enabled=cfg.AMP_ENABLED)
        sched_t.step(val_metrics['dice'])
    
        history_s1['train_loss'].append(train_loss)
        for k, v in val_metrics.items():
            if k == 'thr':  # skip threshold key (not in history dict)
                continue
            history_s1[f'val_{k}'].append(v)
    
        msg = _fmt_epoch("S1", ep+1, cfg.EPOCHS_STAGE1, train_loss, val_metrics)
    
        if val_metrics['dice'] > best_dice_s1:
            best_dice_s1 = val_metrics['dice']
            best_epoch_s1 = ep + 1
            best_metrics_s1 = val_metrics.copy()   # save all metrics at this epoch
            torch.save(model_t.state_dict(), os.path.join(cfg.OUTPUT_DIR, "best_teacher_final.pth"))
            print(msg + "  >> Model Saved with best dice score!")
        else:
            print(msg)



    print("\n" + "="*40 + "\n   BEST TEACHER PERFORMANCE (VAL)\n" + "="*40)
    print(f" - Best Epoch: {best_epoch_s1}")
    for k, v in best_metrics_s1.items():
        if k != "thr":
            print(f" - Val {k.replace('_',' ').title():<15}: {v:.4f}")
    print("="*40)

    plot_history(
    history_s1,
    stage_name="Final Teacher Training",
    save_dir=os.path.join(Config.OUTPUT_DIR, "history_stage1"),
    loss_axis_mode="zero",       
    loss_min_range=0.10,         
    loss_pad_frac=0.10,
    clamp_floor_to_zero=True,
    dpi=500)

    # -------- Stage 2: Pseudo-labels (unlabeled only) --------
    print("\n Final Run : Stage 2 (Pseudo-Label Generation) ")
    cfg.PSEUDO_LABEL_DIR = os.path.join(cfg.OUTPUT_DIR, "pseudo_masks_final")
    os.makedirs(cfg.PSEUDO_LABEL_DIR, exist_ok=True)

    teacher_final = UNet(9,1,cfg.DROPOUT_PROB).to(cfg.DEVICE)
    teacher_final.load_state_dict(torch.load(os.path.join(cfg.OUTPUT_DIR, "best_teacher_final.pth"), map_location=cfg.DEVICE))

    pseudo_loader = DataLoader(
        Brats2_5D_Dataset(unlabeled_files_all, cfg.PROCESSED_DIR, return_fname=True),
        batch_size=cfg.BATCH_SIZE, shuffle=False, **dl_kwargs(cfg)
    )
    generate_pseudo_masks(teacher_final, pseudo_loader, cfg, cfg.PSEUDO_LABEL_DIR)

    

    # -------- Stage 3: Student (labeled + pseudo) --------

    
    del model_t, teacher_final, train_loader, val_loader, pseudo_loader
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    print("\n Final Run : Stage 3 (Student)")
    labeled_suffixes     = {pid.split('_')[-1] for pid in labeled_ids}
    pseudo_files_exist   = [f for f in unlabeled_files_all if os.path.exists(os.path.join(cfg.PSEUDO_LABEL_DIR, f))]
    combined_files       = labeled_files_all + pseudo_files_exist
    combined_ds          = CombinedDataset(combined_files, cfg, labeled_suffixes)
    sampler_s            = create_sampler_for_student_fast(combined_files, cfg.PSEUDO_LABEL_DIR, labeled_suffixes, neg_ratio)
    combined_loader      = DataLoader(combined_ds, batch_size=cfg.BATCH_SIZE, sampler=sampler_s, **dl_kwargs(cfg))

    student  = UNet(9,1,cfg.DROPOUT_PROB).to(cfg.DEVICE)
    student.load_state_dict(torch.load(os.path.join(cfg.OUTPUT_DIR, "best_teacher_final.pth"), map_location=cfg.DEVICE))
    optim_s  = optim.AdamW(student.parameters(), lr=cfg.LEARNING_RATE/2, weight_decay=cfg.WEIGHT_DECAY)
    sched_s  = ReduceLROnPlateau(optim_s, 'max', patience=cfg.SCHEDULER_PATIENCE, verbose=True)
    loss_s   = StudentLoss(cfg.S3_BCE_WEIGHT, cfg.S3_DICE_WEIGHT, cfg.S3_HAUSDORFF_WEIGHT, cfg.STUDENT_TEMP, cfg.STUDENT_BOOTSTRAP_BETA)
    scaler_s = torch.amp.GradScaler(enabled=cfg.AMP_ENABLED)

    history_s2 = {k:[] for k in ['train_loss','val_loss','val_dice','val_hd95','val_neg_acc','val_precision','val_recall']}

    student_val_loader = DataLoader(
        Brats2_5D_Dataset(val_files_all, cfg.PROCESSED_DIR),
        batch_size=cfg.BATCH_SIZE, shuffle=False, **dl_kwargs(cfg)
    )

    best_dice_s2 = -1.0
    best_epoch_s2 = -1
    best_metrics_s2 = None
    
    for ep in range(cfg.EPOCHS_STAGE2):
        train_loss  = train_one_epoch(
            combined_loader, student, optim_s, loss_s, scaler_s,
            cfg.DEVICE, cfg.GRAD_CLIP_NORM, amp_enabled=cfg.AMP_ENABLED
        )
        val_metrics = evaluate_model(student_val_loader, student, loss_s, cfg.DEVICE, amp_enabled=cfg.AMP_ENABLED)
        sched_s.step(val_metrics['dice'])
    
        history_s2['train_loss'].append(train_loss)
        for k, v in val_metrics.items():
            if k == 'thr':  # skip threshold key
                continue
            history_s2[f'val_{k}'].append(v)
    
        msg = _fmt_epoch("S3", ep+1, cfg.EPOCHS_STAGE2, train_loss, val_metrics)
    
        if val_metrics['dice'] > best_dice_s2:
            best_dice_s2 = val_metrics['dice']
            best_epoch_s2 = ep + 1
            best_metrics_s2 = val_metrics.copy()   # save all metrics at this epoch
            torch.save(student.state_dict(), os.path.join(cfg.OUTPUT_DIR, "best_student_final.pth"))
            print(msg + "  >> Model Saved with best dice score!")
        else:
            print(msg)



    print("\n" + "="*40 + "\n   BEST STUDENT PERFORMANCE (VAL)\n" + "="*40)
    print(f" - Best Epoch: {best_epoch_s2}")
    for k, v in best_metrics_s2.items():
        if k != "thr":
            print(f" - Val {k.replace('_',' ').title():<15}: {v:.4f}")
    print("="*40)

    plot_history(
    history_s2,
    stage_name="Final Student Training ",
    save_dir=os.path.join(Config.OUTPUT_DIR, "history_stage3"),
    loss_axis_mode="zero",   
    loss_min_range=0.25,         
    loss_pad_frac=0.10,          
    clamp_floor_to_zero=True,
    dpi=500)

    # -------- Final evaluation on FULL test set (with val-chosen threshold) --------
    print("\n Final Evaluation on Hold-Out Test Set ")
    device_eval = cfg.DEVICE
    final_model = UNet(9,1,cfg.DROPOUT_PROB).to(device_eval)
    final_model.load_state_dict(torch.load(os.path.join(cfg.OUTPUT_DIR, "best_student_final.pth"), map_location=device_eval))

    # 1) Choose threshold on validation (sweep happens inside; VRAM-safe)
    val_choice = evaluate_model(student_val_loader, final_model, loss_s, device_eval, amp_enabled=cfg.AMP_ENABLED)
    best_thr = float(val_choice.get('thr', 0.5))

    # 2) Test at the chosen threshold (no sweep on test)
    test_loader_full = DataLoader(
        Brats2_5D_Dataset(test_files_all, cfg.PROCESSED_DIR),
        batch_size=cfg.BATCH_SIZE, shuffle=False, **dl_kwargs(cfg)
    )
    test_metrics = evaluate_model(test_loader_full, final_model, loss_s, device_eval, thr=best_thr, amp_enabled=cfg.AMP_ENABLED)
    print("\n" + "="*40 + "\n   FINAL TEST SET PERFORMANCE\n" + "="*40)
    for k, v in test_metrics.items():
        print(f" - Test {k.replace('_',' ').title():<15}: {v:.4f}")
    print("="*40)

    # Save metrics + chosen threshold
    test_metrics_out = dict(test_metrics)
    test_metrics_out["thr_used"] = best_thr
    with open(os.path.join(cfg.OUTPUT_DIR, "final_test_metrics.json"), 'w') as f:
        json.dump(test_metrics_out, f, indent=2)

    # -------- Save 10x4 grid (VAL) using the same threshold --------
    val_loader_full_with_names = DataLoader(
        Brats2_5D_Dataset(val_files_all, cfg.PROCESSED_DIR, return_fname=True),
        batch_size=1, shuffle=False, **dl_kwargs(cfg)
    )
    teacher_for_pseudo = UNet(9,1,cfg.DROPOUT_PROB).to(device_eval)
    teacher_for_pseudo.load_state_dict(torch.load(os.path.join(cfg.OUTPUT_DIR, "best_teacher_final.pth"), map_location=device_eval))

    grid_path = os.path.join(cfg.OUTPUT_DIR, "final_examples_grid_val.png")
    
    save_final_visuals_grid(
        pred_model=final_model,
        teacher_model_for_pseudo=teacher_for_pseudo,
        loader_with_fnames=val_loader_full_with_names,
        device=device_eval,
        config=cfg,
        num_examples=10,
        save_path=grid_path,
        thr=best_thr
    )




###  FINAL EXECUTION PIPELINE




def run_pipeline(config):
    
    print("Starting To Run Pipeline")
    source_params_path = '/kaggle/input/optuna-traning-params/best_params.json'
    # Destination is the writable output directory
    dest_params_path = os.path.join(config.OUTPUT_DIR, "best_params.json")

    if os.path.exists(source_params_path):
        print(f"\n[Step 1] Pre-tuned parameters found at '{source_params_path}'.")
        print("SKIPPING Optuna Tuning.")
        
        # Ensure output directory exists and copy the parameters file
        try:
            os.makedirs(config.OUTPUT_DIR, exist_ok=True)
            shutil.copy(source_params_path, dest_params_path)
            if os.path.exists(dest_params_path):
                 print(f"Successfully copied parameters to '{dest_params_path}'.")
            else:
                 # If copy fails, raise an error as this is unexpected
                 raise IOError("Failed to copy parameters file.")
        except Exception as e:
            print(f"ERROR: Could not copy parameters from '{source_params_path}'. Error: {e}")
            print("Attempting to run tuning as a fallback...")
            run_tuning(config)

    else:
        print(f"\n[Step 1] Pre-tuned parameters NOT found at '{source_params_path}'.")
        print("-> STARTING Optuna hyperparameter tuning.")
        run_tuning(config)
        print("Optuna tuning complete.")


    # --- Memory Cleanup ---
    print("\nCleaning up memory before final training...")
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        
    print("\n[Step 2] Proceeding to final training and evaluation...")
    run_final_training_single(config)

    print("\nPipeline Finished")


run_pipeline(Config)
```
